# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
#categorical plots
#<1>bar plots
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
#x=categorical column
#y=numerical column
#aggregate of groupby.....
```

```python
#<2>count plot
#here estimator counts numbers
sns.countplot(x='sex',data=tips)
```

```python
#<3>box plot
#shows distribution of quantitative data in a way that facilitates comparisn b/w variables
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#x=categorical
#y=numerical
#points outside whiskers are determined to be outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#<4>violin plot
#similar to box plots
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#violin plot unlike box plot allows to plot all the components that corresponds to actual datapoints
#and shows the actual kernel density estimation(KDE) of the underlying distribution
#which as a result reveals more information about the datapoints themselves....
#violin plot shows a l'll more information as compared to box plot...and sometimes these information is limited if only box plot is used.....
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#<5>stip plot
#plots scatter plot based of a category 
#may be hard too interpret as compared to a box plot or a violin plot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

```python
#<6>Swarm plot
#combining scatter plot/strip plot v/s violin plot creates a speciall type of plot known as swarm plot
#swarm plot is very similar to strip plot except the points are adjusted in such a way so that they don't overlap....which as a result gives a better representation of the distribution of values...
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
#disadvantage of swarm plot is that it sometimes don't scale that well to very large numbers both in terms of ability to show all the points (points just go too wide) and also takes a lot of computation to begin to arrange them.....
#swarm plots are not meant for very large datasets....
```

```python
#swarm plots can be combines with violin plot to show both pieces of information stacked onn top of each other....
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
#the above piece of analysis/plots is not much suitable for presentation but just for exploratory data analysis....
#helps to immediately understand data
#for presentation box plots and count plots/bar plots are much more convinient..... 
#swarm plot,violin plot,strip plots are more suitable for analysis and is much useful for exploring the data...
```

```python
#<7>Factor plot
#factor plot is the most general form of all these plots
#it takes all the above parameters as its argument...
#but it also takes a kind argument which describes the factor plot.....
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
