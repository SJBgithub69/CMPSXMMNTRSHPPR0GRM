```python
#creating a function that takes batsman name as input as a sting and returns with the value against which teams that particular batsman has scored maximum runs....
```

```python
def runs_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
runs_scored('V Kohli')
```
