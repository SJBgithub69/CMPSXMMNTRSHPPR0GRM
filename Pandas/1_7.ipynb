```python
#pivot table oo IPL dataset
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery.head()
```

```python
#writing a code to create a pivot table:-
    #columns=team name
    #rows=overs(1-20)
    #values to be plotted inside-sixes scored by every team in particular overs
#ex:-no. of sixes scored by CSK and MI in nth and mth over respectively.
```

```python
#refining data
#only those columns are required where batsman shot six...

mask=delivery['batsman_runs']==6
six=delivery[mask]
six.head()
```

```python
six.shape
#total no. of sixes shot in IPL
```

```python
#plotting pivot table on the new dataframe thus obtained...
six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
#converting pivot table into a heat map
    #porting the values in the pivot table to a heat map....
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
import seaborn as sns
sns.heatmap(pt)
#heatmap() is a function in seaborn
```
