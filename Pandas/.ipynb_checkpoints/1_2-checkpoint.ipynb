```python
#The groupby() function
#Fortune 500 dataset
import numpy as np
import pandas as pd
```

```python
company=pd.read_csv('Fortune 500.csv')
```

```python
company
```

```python
company.head()
```

```python
#the groupby() funtion
#data analysis can also be done by creating groups
#groups can be created on the basis of categorical data i.e the series/column that is repeating for 2 or more rows
#in this case(for this dataset) groups can be created on the basis of either sector or industry
```

```python
#creating a groupby object from this dataframe by creating groups using the sector coulumn
company.groupby('Sector')
```

```python
sectors=company.groupby('Sector')
```

```python
sectors.first()
#.first() function displays first items in each category or group created and thus getting the same number of rows as that of no. of groups
```

```python
#complimentary function for .first() is .last()
sectors.last()
#lists last items in each category
```

```python
sectors.groups
#.groups is an attribute
#a dictionary{} is returned in which key is name of the sector and its valus is a list that is returned with the id of each company
#.groups attribute returns a dictionary of which every key represents a group and its value is returned with every member of that group with its index and can be cross-checked using the iloc function....
```

```python
company.iloc[23, :]
#: at the columns part/section prints out all the columns
```

```python
sectors.get_group('Energy')
#get_group() function returns with all the values under a particular group
#get_group is very useful when a particular group/item is to be fetched from a groupby object....and inside parenthesis the name of the group is given as a string...
```

```python
 sectors.get_group('Apparel')
```

```python
sectors.sum()
#numerical columns are summed up in the dataframe or in this case all the groups in groupby function.....
```
