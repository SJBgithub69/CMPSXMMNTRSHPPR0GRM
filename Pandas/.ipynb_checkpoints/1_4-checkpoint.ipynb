```python
#isin() function
#Ques=Find the most destructive death over batsman in the history of IPL
import pandas as pd
import numpy as np
```

```python
#on the basis of strike rate
#strike rate is basically number of runs scored by batsman in 100 balls..
#strike rate=(Number of runs scored/number of balls played)*100
#condition-batsman played minimum 200 balls between overs 16-20.....
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery
```

```python
delivery.groupby('batsman')['batsman_runs'].count()
#equals no. of balls played=no. of rows
```

```python
#150000 rows in total in this dataframe and all are for overs ranging from 1 to 20....
#acc. to condition:-over must be death over i.e b/w 16-20....
delivery['over']>15
```

```python
delivery[delivery['over']>15]
#death over=over>15
```

```python
death_over=delivery[delivery['over']>15]
death_over.groupby('batsman')['batsman_runs'].count()
#player that played in death over:---
```

```python
death_over.groupby('batsman')['batsman_runs'].count()>200
#boolean series obtained....
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
all_batsman
```

```python
all_batsman>200
```

```python
x=all_batsman>200
all_batsman[x]
#boolean indexing
#only those batsman/ids are obtained whose value is true
```

```python
 all_batsman[x].shape
```

```python
 all_batsman[x].index
```

```python
#tolist() function :-converts index object into list
```

```python
batsman_list=all_batsman[x].index.tolist()
batsman_list
```

```python
#calculating strike rate of these batsman obtained above.....
#then max strike rate can be obtained....
```

```python
#for calculating strike rate:-
    #runs scored by all these 43 batsman
    #balls played by these 43 batsman
delivery
```

```python
#here masking cannot be done since no. of data is more and cannot be individually handled....
#like there has to be 43 operations to be performed like below..
    #delivery['batsman']=='V Kohli' ....
#here isin() function comes in role....
delivery['batsman']
```

```python
delivery['batsman'].isin(batsman_list)
```

```python
delivery[delivery['batsman'].isin(batsman_list)]
```

```python
delivery[delivery['batsman'].isin(batsman_list)].shape
#balls played by these 43 player
#total balls played = 66006
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
final.groupby('batsman')['batsman_runs'].sum()
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```
