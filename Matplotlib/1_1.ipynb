# Pie Chart

```python
#creating pie chart to invest on various sectors of a company
#budget
```

```python
areas=['Marketing','Sales','Development','HR','Customer Support']
budget=[2.5,4,10,1,6]
```

```python
plt.pie(budget, labels=areas)
```

```python
plt.pie(budget, labels=areas)
plt.show()
```

```python
plt.pie(budget, labels=areas,autopct="%0.1f%%")
plt.show()
#for showing %age autopct="%0.1f%%"
```

```python
plt.pie(budget, labels=areas,autopct="%0.4f%%")
plt.show()
#for showing %age autopct="%0.4f%%" upto 4 decimal places
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,656,899]
chor_bazaar=[100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id,online_price)
#2D plot for lists or Numpy arrays
```

```python
plt.plot(product_id,online_price)
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
```

```python
plt.plot(product_id,online_price)
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
#for not showing the autogenerated text/technical info above the graph..
```

```python
plt.plot(product_id,online_price, color='black')
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
```

```python
plt.plot(product_id,online_price, color='black',marker='+')
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
```

```python
plt.plot(product_id,online_price, color='black',marker='D')
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
#d=diamond
```

```python
plt.plot(product_id,online_price, color='black',marker='D',markersize=6)
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
```

```python
plt.plot(product_id,online_price, color='red',marker='+',markersize=5)
plt.plot(product_id,offline_price, color='green',marker='.',markersize=5)
plt.plot(product_id,chor_bazaar, color='black',marker='D',markersize=5)

plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
```

```python
plt.plot(product_id,online_price, color='red',marker='+',markersize=5,label="Online Prices")
plt.plot(product_id,offline_price, color='green',marker='.',markersize=5,label="Offline Prices")
plt.plot(product_id,chor_bazaar, color='black',marker='D',markersize=5,label="Chor Bazaar Prices")

plt.legend()
#for labelling each and every curve
#plt.legend() for validating label attribute

plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparisn")
plt.show()
```
